```python
# MACHINE LEARNING
# Step 1 :splitting DAta sets into Features (x)and Outcomes (y)


array =data.values
array 

X = array[0:50, 0:8]
Y  = array[0:50, 8]
Y
```

```python
# step 2 :Spiliting the Dateset(x,y) into Training set (70%) and testing set (30%)
# X -> X-train x-train,X_ test
# Y-> Y_train, Y_test
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.3, random_state = 42)
```

```python
# step3: TRAINING PHASE (70%) -X_train, Y_train
# Scientific e.g Decision Tree, K Neghbors ,SVM logistics Regression...

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB 
```

```python
# step4: CROSS VALIDATION 
# Technique to pick the best model for a specific problem
models = []

models.append(('Decision Tree',DecisionTreeClassifier()))
models.append(('KN',KNeighborsClassifier()))
models.append(('Gradient boosting',GradientBoostingClassifier()))
# models.append(('Logistic Regression', LogisticRegression))
models.append(('Gaussian',GaussianNB()))
models.append(('Support vector',SVC(gamma='auto')))
models.append(('Random forest',RandomForestClassifier()))
from sklearn.model_selection import KFold,cross_val_score
for name, model  in models:
    kfold = KFold(n_splits =10 ,random_state = 42,shuffle =True)
    results = cross_val_score( model ,X_train,Y_train,cv = kfold, scoring='accuracy')

    print('Model name:',name,'Results',results.mean())
models
```

```python
# step 5:TRAINING 
model =GradientBoostingClassifier ()
model.fit(X_train,Y_train)
```

```python
# step6 :TESTING (testing data)
predictions = model.predict(X_test)
print("Model predicted:", predictions)
```

```python
print("Correct Results:",Y_test)
```

```python
# step7 :model Evaluation
# AS -> Ration of model predictions againist correct Predictions
from sklearn.metrics import accuracy_score
print("Model Accuracy:",accuracy_score(predictions, Y_test)* 100)
```

```python
# PREDICTIONS/DEVELOPMENT
ingredients = [ [30,30,15,15,15,10,15,5] ]
cake = model.predict(ingredients)
print("100% that you are Baking:",cake)
```
